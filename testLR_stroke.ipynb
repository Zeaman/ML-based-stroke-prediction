```python
# Import libraries
import pickle
import numpy as np
import pandas as pd
import tkinter as tk
from tkinter import ttk
from sklearn.preprocessing import StandardScaler
```

```python
# Load the trained stroke model from the pickle file
with open('bestLR_model.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
# Create the GUI
root = tk.Tk()
# Define the title
root.title("Stroke Prediction System")
# Define the size
root.geometry("300x400")
# Create a style object for the widgets
style = ttk.Style()
style.theme_use("clam")
# Configure the style for the labels
style.configure("TLabel", foreground="black", background="white", font=("Helvetica", 12), padding=10)
# Configure the style for the buttons
style.configure("TButton", foreground="white", background="blue", font=("Helvetica", 12), padding=10)


# Gender
gender_label = tk.Label(root, text='Gender:')
gender_label.grid(row=0, column=0)

# Create a list of encoded values for gender
gender_values = [ "Female", "Male", "Other"]

# Create a string variable to store the selected value
gender_var = tk.StringVar()
gender_var.set(gender_values[0])  # Set the default value

# Create the dropdown menu
gender_dropdown = tk.OptionMenu(root, gender_var, *gender_values)
gender_dropdown.grid(row=0, column=1)

# Function to convert the selected value to integer
def get_gender():
    gender = gender_values.index(gender_var.get())
    return gender

# Age
age_label = tk.Label(root, text='Age:')
age_label.grid(row=1, column=0)
age_entry = tk.Entry(root)
age_entry.grid(row=1, column=1)

# Hypertension
hypertension_label = tk.Label(root, text='Hypertension:')
hypertension_label.grid(row=2, column=0)

# Create a list of encoded values for gender
hypertension_values = [ "Yes", "No"]

# Create a string variable to store the selected value
hypertension_var = tk.StringVar()
hypertension_var.set(hypertension_values[0])  # Set the default value

# Create the dropdown menu
hypertension_dropdown = tk.OptionMenu(root, hypertension_var, *hypertension_values)
hypertension_dropdown.grid(row=2, column=1)

# Function to convert the selected value to integer
def get_hypertension():
    hypertension = hypertension_values.index(hypertension_var.get())
    return hypertension

# Heart Disease
heart_disease_label = tk.Label(root, text='Heart Disease:')
heart_disease_label.grid(row=3, column=0)

# Create a list of encoded values for gender
heart_disease_values = [ "Yes", "No"]

# Create a string variable to store the selected value
heart_disease_var = tk.StringVar()
heart_disease_var.set(heart_disease_values[0])  # Set the default value

# Create the dropdown menu
heart_disease_dropdown = tk.OptionMenu(root, heart_disease_var, *heart_disease_values)
heart_disease_dropdown.grid(row=3, column=1)

# Function to convert the selected value to integer
def get_heart_disease():
    heart_disease = heart_disease_values.index(heart_disease_var.get())
    return heart_disease

# Ever Married
ever_married_label = tk.Label(root, text='Ever Marriged:')
ever_married_label.grid(row=4, column=0)

# Create a list of encoded values for gender
ever_married_values = [ "Yes", "No"]

# Create a string variable to store the selected value
ever_married_var = tk.StringVar()
ever_married_var.set(ever_married_values[0])  # Set the default value

# Create the dropdown menu
ever_married_dropdown = tk.OptionMenu(root, ever_married_var, *ever_married_values)
ever_married_dropdown.grid(row=4, column=1)

# Function to convert the selected value to integer
def get_ever_married():
    ever_married = ever_married_values.index(ever_married_var.get())
    return ever_married

# Work Type
work_type_label = tk.Label(root, text='Work Type:')
work_type_label.grid(row=5, column=0)

# Create a list of encoded values for gender
work_type_values = [ "Govt_job", "Never_worked", "Private", "Never_worked", "Self-employed", "Children"]

# Create a string variable to store the selected value
work_type_var = tk.StringVar()
work_type_var.set(work_type_values[0])  # Set the default value

# Create the dropdown menu
work_type_dropdown = tk.OptionMenu(root, work_type_var, *work_type_values)
work_type_dropdown.grid(row=5, column=1)

# Function to convert the selected value to integer
def get_work_type():
    work_type = work_type_values.index(work_type_var.get())
    return work_type

# Residence Type
residence_type_label = tk.Label(root, text='Residence Type:')
residence_type_label.grid(row=6, column=0)

# Create a list of encoded values for gender
residence_type_values = [ "Rural", "Urban"]

# Create a string variable to store the selected value
residence_type_var = tk.StringVar()
residence_type_var.set(residence_type_values[0])  # Set the default value

# Create the dropdown menu
residence_type_dropdown = tk.OptionMenu(root, residence_type_var, *residence_type_values)
residence_type_dropdown.grid(row=6, column=1)

# Function to convert the selected value to integer
def get_residence_type():
    residence_type = residence_type_values.index(residence_type_var.get())
    return residence_type

# Glucose Level
glucose_level_label = tk.Label(root, text='Glucose Level:')
glucose_level_label.grid(row=7, column=0)
glucose_level_entry = tk.Entry(root)
glucose_level_entry.grid(row=7, column=1)

# BMI
bmi_label = tk.Label(root, text='BMI:')
bmi_label.grid(row=8, column=0)
bmi_entry = tk.Entry(root)
bmi_entry.grid(row=8, column=1)

# Smoking Status
smoking_status_label = tk.Label(root, text='Smoking Status:')
smoking_status_label.grid(row=9, column=0)

# Create a list of encoded values for gender
smoking_status_values = [ "Unknown ", "Formerly smoked", "Never smoked", "Smokes"]

# Create a string variable to store the selected value
smoking_status_var = tk.StringVar()
smoking_status_var.set(smoking_status_values[0])  # Set the default value

# Create the dropdown menu
smoking_status_dropdown = tk.OptionMenu(root, smoking_status_var, *smoking_status_values)
smoking_status_dropdown.grid(row=9, column=1)

# Function to convert the selected value to integer
def get_smoking_status():
    smoking_status = smoking_status_values.index(smoking_status_var.get())
    return smoking_status

# Submit button
def predict_stroke():
    # Get the input values
    age = float(age_entry.get())
    gender = get_gender()
    hypertension = get_hypertension()
    heart_disease = get_heart_disease()
    ever_married = get_ever_married()
    work_type = get_work_type()
    residence_type = get_residence_type()
    glucose_level = float(glucose_level_entry.get())
    bmi = float(bmi_entry.get())
    smoking_status = get_smoking_status()

    # Prepare the input as a numpy array
    input_data = np.array([[age, gender, hypertension, heart_disease, ever_married, work_type, residence_type, glucose_level,smoking_status, bmi]])

    # Predict the stroke probability
    stroke_prob = model.predict_proba(input_data)[:, 1][0]

    # Display the result
    if stroke_prob >= 0.5:
        result_label.configure(text=f'Stroke Predicted: Yes ({stroke_prob*100:.2f}%)')
    else:
        result_label.configure(text=f'Stroke Predicted: No ({(1-stroke_prob)*100:.2f}%)')

submit_button = tk.Button(root, text='Submit', command=predict_stroke,font=("Helvetica", 10))
submit_button.grid(row=13, columnspan=2)

# Result label
result_label = tk.Label(root, text='Stroke Predicted:',font=("Helvetica", 11))
result_label.grid(row=11, columnspan=3)

root.mainloop()
```

## The values you need to provide are:

-Gender: your gender, represented as a binary value (0 for female, 1 for male, 2 for other)\
-Age: your age in years (a positive number)\
-Hypertension: whether you have hypertension or not, represented as a binary value (0 for no, 1 for yes)\
-Heart Disease: whether you have heart disease or not, represented as a binary value (0 for no, 1 for yes)\
-Ever Married: whether you have ever been married or not, represented as a binary value (0 for no, 1 for yes)\
-Work Type: your work type, represented as an integer (0 for "Govt_job", 1 for "Never_worked", 2 for "Private", 3 for "Self-employed", 4 for "children")\
-Residence Type: your residence type, represented as an integer (0 for "Rural", 1 for "Urban")\
-Glucose Level: your glucose level in mg/dL (a positive number)\
-BMI: your body mass index (BMI) in kg/m^2 (a positive number)\
-Smoking Status: your smoking status, represented as an integer (0 for "unknown ", 1 for "formerly smoked", 2 for "never smoked", 3 for "smokes")
